```python
# splitting the cabin column into more useful bits

cabin_data = train_df[~train_df["Cabin"].isna()].copy()
cabin_data["cabin_deck"] = cabin_data["Cabin"].str.split("/").str[0]
cabin_data["cabin_num"] = cabin_data["Cabin"].str.split("/").str[1]
cabin_data["cabin_side"] = cabin_data["Cabin"].str.split("/").str[2]
cabin_data
```

```python
cabin_data['cabin_deck'].nunique(), cabin_data['cabin_num'].nunique(), cabin_data['cabin_side'].nunique()
```

```python
sns.countplot(x="cabin_deck", hue="Transported", data=cabin_data)
plt.show()
```

```python
sns.countplot(x="cabin_side", hue="Transported", data=cabin_data)
plt.show()
```

```python
def map_to_bins(num):
    num = int(num)
    if num <= 300:
        return 1
    elif num <= 600:
        return 2
    elif num <= 900:
        return 3
    elif num <= 1200:
        return 4
    elif num <= 1500:
        return 5
    else:
        return 6
```

```python
cabin_data["cabin_num_bin"] = cabin_data["cabin_num"].map(map_to_bins)
```

```python
sns.countplot(x="cabin_num_bin", hue="Transported", data=cabin_data)
plt.show()
```

```python
train_df_ns = train_df[train_df["CryoSleep"] == False].copy()
```

```python
train_df_ns["total_spending"] = train_df_ns[["RoomService", "FoodCourt", "ShoppingMall", "Spa", "VRDeck"]].apply(sum, axis=1)
```

```python
train_df_ns
```

```python
train_df_ns["total_spending"].describe()
```

```python
def bin_spending(sum):
    if sum <= 865:
        return 1
    elif sum <= 1730:
        return 2
    elif sum <= 2595:
        return 3
    elif sum <= 3460:
        return 4
    elif sum <= 4325:
        return 5
    elif sum <= 5190:
        return 6
    elif sum <= 6055:
        return 7
    elif sum <= 6920:
        return 8
    else:
        return 9
```

```python
train_df_ns["total_spending_bin"] = train_df_ns["total_spending"].map(bin_spending)
```

```python
train_df_ns
```

```python
sns.countplot(x="total_spending_bin", hue="Transported", data=train_df_ns)
plt.show()
```
