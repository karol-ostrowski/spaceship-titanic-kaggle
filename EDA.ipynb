```python
def map_to_bins(num):
    num = int(num)
    if num <= 300:
        return 1
    elif num <= 600:
        return 2
    elif num <= 900:
        return 3
    elif num <= 1200:
        return 4
    elif num <= 1500:
        return 5
    else:
        return 6
```

```python
cabin_data["cabin_num_bin"] = cabin_data["cabin_num"].map(map_to_bins)
```

```python
sns.countplot(x="cabin_num_bin", hue="Transported", data=cabin_data)
plt.title("Impact of cabin number on transportation")
plt.show()
```

```python
# checking the impact of total spending on transpotration
train_df_ns = train_df[train_df["CryoSleep"] == False].copy()
```

```python
train_df_ns["total_spending"] = train_df_ns[["RoomService", "FoodCourt", "ShoppingMall", "Spa", "VRDeck"]].apply(sum, axis=1)
```

```python
train_df_ns
```

```python
train_df_ns["total_spending"].describe()
```

```python
def bin_spending(sum):
    if sum <= 865:
        return 1
    elif sum <= 1730:
        return 2
    elif sum <= 2595:
        return 3
    elif sum <= 3460:
        return 4
    elif sum <= 4325:
        return 5
    elif sum <= 5190:
        return 6
    elif sum <= 6055:
        return 7
    elif sum <= 6920:
        return 8
    else:
        return 9
```

```python
train_df_ns["total_spending_bin"] = train_df_ns["total_spending"].map(bin_spending)
```

```python
train_df_ns
```

```python
sns.countplot(x="total_spending_bin", hue="Transported", data=train_df_ns)
plt.title("Impact of total spending on transportation")
plt.show()
```

```python
# checking how many passenger groups there are
train_df["PassengerGroup"] = train_df["PassengerId"].str.split("_").str[0]
train_df["PassengerGroup"].nunique()
```

```python
sleeping["PassengerGroup"] = sleeping["PassengerId"].str.split("_").str[0]
not_sleeping["PassengerGroup"] = not_sleeping["PassengerId"].str.split("_").str[0]
```

```python
# creating a column that specifies the group size a passenger is travelling in
train_df["group_size"] = train_df.groupby("PassengerGroup")["PassengerGroup"].transform("count")
train_df["group_size"].nunique()
```

```python
sleeping["group_size"] = sleeping.groupby("PassengerGroup")["PassengerGroup"].transform("count")
not_sleeping["group_size"] = not_sleeping.groupby("PassengerGroup")["PassengerGroup"].transform("count")
```

```python
# impace of group size on transportation
sns.countplot(x="group_size", hue="Transported", data=train_df)
plt.title("Impact of group size a passenger in travelling in on transportation")
plt.show()
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 5))
sns.countplot(x="group_size", hue="Transported", data=not_sleeping, ax=axes[0])
axes[0].set_ylim(0, 3000)
axes[0].set_title("Impact of group size a passenger\nis travelling in on transportation\nfor awake passengers")
sns.countplot(x="group_size", hue="Transported", data=sleeping, ax=axes[1])
axes[1].set_ylim(0, 3000)
axes[1].set_title("Impact of group size a passenger\nis travelling in on transportation\nfor passengers in CryoSleep")
plt.show()
```
